```python
#  Function to find relative bills for each of User in particular category tag

def bill(Name, tag):
    
    ans = (dataset_orig.loc[(dataset_orig['TransactionUserName'] == Name) & (dataset_orig['tags'] == tag)])
    
    
    ans.sort_values(by = ['Date'],inplace = True, ascending = False)
    
    ans.index = range(ans.shape[0])
  
    
    return(ans)
    
```

```python
# Function to find latest month bill for each tag corrsponding to each user
# T is a database from above finction

def monthlybill(T):
    a = 0
    i = 0
    while(T['Date'].loc[0].strftime('%m') == T['Date'].loc[i].strftime('%m')):
        a = a + T['Amount'].iloc[i]
        i = i+1
    return(a)
    
    
```

```python
# Function to find recent n month bills for each user corresponding to each tag

def recentmonthsbill(T,n):
    A = []
    i = 0
    k = 0
    for j in range(n):
        a = 0
        while(T['Date'].loc[k].strftime('%m') == T['Date'].loc[i].strftime('%m')):
           
            a = a + T['Amount'].iloc[i]
            i = i+1
        k = i
        A.append(a)
    return(A)
```

```python
recentmonthsbill(X,4)
```

```python
# For Impulsive Shopper

Z = bill('       VIBHORE GOYAL','Shopping')
```

```python
# Function to find out purchase frequency for n months for each use corresponding to each tag

def monthpurchasefrequency(T,n):
    A = []
    i = 0
    k = 0
    for j in range(n):
        l = 0
        while(T['Date'].loc[k].strftime('%m') == T['Date'].loc[i].strftime('%m')):
           
            l = l+1
            i = i+1
        k = i
        A.append(l)
    return(A)
```

```python
# Function to categorise into Impulsive or Non Impulsive shopper

def impulsiveshopper(K):
    for i in K:
        if(i>10):
            return("Yes")

    return("No")
```
